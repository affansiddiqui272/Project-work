```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/content/global_co2.csv')
print(data.head())
```

```python
plt.figure(figsize=(10, 6))
plt.plot(data['Year'], marker='1', label='Historical Data')
plt.title('CO2 Production per Capita (1970-2010)')
plt.xlabel('Year')
plt.ylabel('CO2 per Capita')
plt.grid()
plt.show()
```

```python
X = data[['Year']]
y = data['Gas Flaring']
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
future_years = np.array([[2011], [2012], [2013]])
predictions = model.predict(future_years)
```

```python
for year, prediction in zip(future_years.flatten(), predictions):
    print(f"Predicted CO2 production per capita for {year}: {prediction:.2f}")
```

```python
plt.figure(figsize=(10, 6))
plt.plot(data['Year'], marker='o', label='Historical Data')
plt.plot(future_years, predictions, 'ro', label='Predictions')
plt.title('CO2 Production per Capita with Predictions')
plt.xlabel('Year')
plt.ylabel('CO2 per Capita')
plt.legend()
plt.grid()
plt.show()
```
